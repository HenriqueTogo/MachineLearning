```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Dataset 2D disponível no [Kaggle](https://www.kaggle.com/datasets/samuelcortinhas/2d-clustering-data?resource=download).

```python
data = np.loadtxt('data_clustering.csv', delimiter=',', dtype=float)
data_label = data[:,2]
data = data[:, 0:2]
plt.title('Plotting Dataset')
plt.plot(data[:,0], data[:,1], 'ro',markeredgecolor=(0,0,0,1))
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
k_clusters = 2
kmeans = KMeans(n_clusters=k_clusters, init='k-means++').fit(data)
predict = kmeans.fit_predict(data)

clusters = [[], [], [], [], [], [], []]

for i in range(len(data)):
    if (predict[i] == 0):
        clusters[0].append(data[i,:])
    elif (predict[i] == 1):
        clusters[1].append(data[i,:])
    elif (predict[i] == 2):
        clusters[2].append(data[i,:])
    elif (predict[i] == 3):
        clusters[3].append(data[i,:])
    elif (predict[i] == 4):
        clusters[4].append(data[i,:])
    elif (predict[i] == 5):
        clusters[5].append(data[i,:])
    elif (predict[i] == 6):
        clusters[6].append(data[i,:])

plt.title(f'Plotting Clustering with {k_clusters} cluster(s).')


markers = ['bo', 'go', 'ro', 'co', 'mo', 'yo', 'ko']

for i in range(k_clusters):
    if (predict[i] == 0):
        plt.plot(clusters[0][:,0], clusters[0][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[0], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 1):
        plt.plot(clusters[1][:,0], clusters[1][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[1], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 2):
        plt.plot(clusters[2][:,0], clusters[2][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[2], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 3):
        plt.plot(clusters[3][:,0], clusters[3][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[3], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 4):
        plt.plot(clusters[4][:,0], clusters[4][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[4], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 5):
        plt.plot(clusters[5][:,0], clusters[5][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[5], markeredgecolor=(0,0,0,1), markersize=12)
    elif (predict[i] == 6):
        plt.plot(clusters[6][:,0], clusters[6][:,1], 'ro',markeredgecolor=(0,0,0,1))
        plt.plot(kmeans.cluster_centers_[i,0], kmeans.cluster_centers_[i,1], markers[6], markeredgecolor=(0,0,0,1), markersize=12)

plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
